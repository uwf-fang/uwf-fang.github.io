## 5.1 Join, filter and select

Task: Join two tables on key column and select name and address fields. Keep records with age greater than 21

```python
# DataFrames preparation
data1 = [
    (1, "Alice", 25),
    (2, "Bob", 20),
    (3, "Charlie", 30)
]
df1 = spark.createDataFrame(data1, ["key", "name", "age"])
# Show df1
print("DataFrame df1:")
df1.show()

data2 = [
    (1, "123 Main St"),
    (2, "456 Elm St"),
    (3, "789 Oak St")
]
df2 = spark.createDataFrame(data2, ["key", "address"])
print("DataFrame df2:")
df2.show()
```

### Use Python API calls to achieve the goal

```python
df1.join(df2, on="key", how="inner") \
    .filter(df1.age > 21) \
    .select('name', 'address') \
    .show()
```

### Use SQL to achieve the goal

```python
# Create internal temporary view for dataframes as tables
df1.createOrReplaceTempView("table1")
df2.createOrReplaceTempView("table2")

# Execute the query
query = """
    SELECT t1.name, t2.address
    FROM table1 t1
    INNER JOIN table2 t2 ON t1.key = t2.key
    WHERE t1.age > 21
"""
spark.sql(query).show()
```
